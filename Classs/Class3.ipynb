```python
# append(): Adds an element to the end of the list.
fruits = ['apple', 'banana', 'cherry']
fruits.append('orang')
print(f"append() example: {fruits}")
```

```python
# extend(): Adds all the elements of an iterable (list, tuple, string etc.) to the end of the list.
numbers = [1, 2, 3]
more_numbers = [4, 5, 6]
numbers.extend(more_numbers)
print(f"extend() example: {numbers}")
```

```python
# remove(): Removes the first item with the specified value.
cars = ['Ford', 'BMW', 'Volvo']
cars.remove('Volvo')
print(f"remove() example: {cars}")
```

```python
# insert(): Adds an element at the specified position.
colors = ['red', 'green', 'blue']
colors.insert(1, 'yellow')
print(f"insert() example: {colors}")
```

```python
# clear(): Removes all the elements from the list.
items = [1, 'a', 2.5]
items.clear()
print(f"clear() example: {items}")
```

```python
# sort(): Sorts the list ascending by default.
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
numbers.sort()
print(f"sort() example: {numbers}")
```

```python
# reverse(): Reverses the order of the list.
letters = ['a', 'b', 'c', 'd']
letters.reverse()
print(f"reverse() example: {letters}")
```

```python
# copy(): Returns a shallow copy of the list.
original = [1, 2, 3]
new_list = original.copy()
new_list.append(4)
print(f"copy() example: original={original}, new_list={new_list}")
```

```python
# count(): Returns the number of elements with the specified value.
numbers = [1, 2, 3, 4, 3, 2, 1]
count_of_3 = numbers.count(3)
print(f"count() example: {count_of_3}")
```

```python
# pop(): Removes the element at the specified position. If no index is specified, it removes and returns the last item.
alphabets = ['a', 'b', 'c', 'd']
popped_element = alphabets.pop(1)
print(f"pop() example: popped_element={popped_element}, list={alphabets}")
```

### List Built-in Functions

```python
# append(): Adds an element to the end of the list.
print("--- append() examples ---")
# 1. Appending a number
numbers = [1, 2, 3]
numbers.append(4)
print(f"1. Appending a number: {numbers}")
# 2. Appending a string
names = ['John', 'Peter']
names.append('Susan')
print(f"2. Appending a string: {names}")
# 3. Appending another list
lists = [[1, 2], [3, 4]]
lists.append([5, 6])
print(f"3. Appending a list: {lists}")
```

### Tuple Built-in Functions
Tuples are immutable, so they don't have methods like `append()` or `remove()` that modify the tuple in place. However, you can use `+` to concatenate tuples and create a new one.

```python
# Tuple concatenation
print("--- Tuple concatenation examples ---")
# 1. Concatenating tuples of numbers
numbers1 = (1, 2, 3)
numbers2 = (4, 5, 6)
combined_numbers = numbers1 + numbers2
print(f"1. Concatenating number tuples: {combined_numbers}")
# 2. Concatenating tuples of strings
names1 = ('John', 'Peter')
names2 = ('Susan',)
combined_names = names1 + names2
print(f"2. Concatenating string tuples: {combined_names}")
# 3. Concatenating mixed-type tuples
mixed1 = (1, 'hello')
mixed2 = (3.14, 'world')
combined_mixed = mixed1 + mixed2
print(f"3. Concatenating mixed-type tuples: {combined_mixed}")
```

### Dictionary Built-in Functions

```python
# keys(): Returns a view object that displays a list of all the keys in the dictionary.
print("--- keys() examples ---")
# 1. Keys of a simple dictionary
person = {'name': 'John', 'age': 30, 'city': 'New York'}
print(f"1. Simple dictionary keys: {person.keys()}")
# 2. Keys of a dictionary with numeric keys
numeric_keys = {1: 'a', 2: 'b', 3: 'c'}
print(f"2. Numeric keys: {numeric_keys.keys()}")
# 3. Keys of a JSON-like dictionary
json_like = {'user': {'id': 123, 'name': 'John Doe'}, 'posts': [{'title': 'Post 1'}, {'title': 'Post 2'}]}
print(f"3. JSON-like dictionary keys: {json_like.keys()}")

# values(): Returns a view object that displays a list of all the values in the dictionary.
print("\n--- values() examples ---")
# 1. Values of a simple dictionary
print(f"1. Simple dictionary values: {person.values()}")
# 2. Values of a dictionary with numeric keys
print(f"2. Numeric key values: {numeric_keys.values()}")
# 3. Values of a JSON-like dictionary
print(f"3. JSON-like dictionary values: {json_like.values()}")

# items(): Returns a view object that displays a list of a dictionary's key-value tuple pairs.
print("\n--- items() examples ---")
# 1. Items of a simple dictionary
print(f"1. Simple dictionary items: {person.items()}")
# 2. Items of a dictionary with numeric keys
print(f"2. Numeric key items: {numeric_keys.items()}")
# 3. Items of a JSON-like dictionary
print(f"3. JSON-like dictionary items: {json_like.items()}")

# popitem(): Removes the last inserted key-value pair.
print("\n--- popitem() examples ---")
# 1. Popping from a simple dictionary
popped_item = person.popitem()
print(f"1. Popped item: {popped_item}, remaining dict: {person}")
# 2. Popping from a dictionary with numeric keys
popped_item = numeric_keys.popitem()
print(f"2. Popped item: {popped_item}, remaining dict: {numeric_keys}")
# 3. Popping from a JSON-like dictionary
popped_item = json_like.popitem()
print(f"3. Popped item: {popped_item}, remaining dict: {json_like}")
```
