```python
print("### Nested Loops and Patterns")

print("\n1. Right-angled triangle:")
for i in range(5):
  for j in range(i + 1):
    print("* ", end="")
  print()

print()

print("\n2. Inverted right-angled triangle:")
for i in range(5, 0, -1):
  for j in range(i):
    print("* ", end="")
  print()
```

## Provide code examples demonstrating `break` with numbers.

### Subtask:
Show how to use `break` within a loop iterating through numbers.


**Reasoning**:
Print a header, then implement the `for` loop with a break statement based on a numerical condition, followed by print statements before and after the loop. Then, implement the `while` loop with a break condition based on a counter and include print statements to show the loop's progress and stopping point.



```python
print("### Using `break` with Numbers")

# Example 1: Using break in a for loop
print("\nFor loop with break:")
for i in range(10):
  print(f"Current number: {i}")
  if i == 5:
    print("Reached 5, breaking loop.")
    break
print("For loop finished.")

# Example 2: Using break in a while loop
print("\nWhile loop with break:")
count = 0
while count < 10:
  print(f"Current count: {count}")
  if count == 7:
    print("Count reached 7, breaking loop.")
    break
  count += 1
print("While loop finished.")
```

## Provide code examples demonstrating `continue` with numbers.

### Subtask:
Show how to use `continue` within a loop iterating through numbers.


**Reasoning**:
I will write a single code block to demonstrate the use of the `continue` statement in both a for loop and a while loop, iterating through numbers and skipping specific iterations based on a condition, as per the instructions.



```python
print("### Using `continue` with Numbers")

# Using continue in a for loop
print("\nFor loop with continue:")
for i in range(10):
  if i % 2 == 0:  # Check if the number is even
    print(f"Skipping even number: {i}")
    continue  # Skip the rest of the current iteration
  print(f"Processing odd number: {i}") # This will only print for odd numbers
print("For loop finished.")

# Using continue in a while loop
print("\nWhile loop with continue:")
count = 0
while count < 10:
  if count % 2 != 0: # Check if the number is odd
    print(f"Skipping odd number: {count}")
    count += 1 # Increment before continue to avoid infinite loop
    continue # Skip the rest of the current iteration
  print(f"Processing even number: {count}") # This will only print for even numbers
  count += 1
print("While loop finished.")
```

```python
print("### Using `break` with Strings")

# Example 1: Using break in a for loop with a string
print("\nFor loop with break and string:")
my_string = "programming"
for char in my_string:
  print(f"Current character: {char}")
  if char == 'g':
    print("Found 'g', breaking loop.")
    break
print("For loop finished.")

# Example 2: Using break in a while loop with a list of strings
print("\nWhile loop with break and list of strings:")
words = ["apple", "banana", "cherry", "date", "fig"]
index = 0
while index < len(words):
  current_word = words[index]
  print(f"Current word: {current_word}")
  if current_word == "cherry":
    print("Found 'cherry', breaking loop.")
    break
  index += 1
print("While loop finished.")
```

## Provide code examples demonstrating `continue` with strings.

### Subtask:
Show how to use `continue` within a loop iterating through strings.


**Reasoning**:
The subtask requires demonstrating the use of `continue` with strings in both `for` and `while` loops, including printing messages to show the flow. I will create a code block that includes a header and then two examples, one for a `for` loop iterating through a string and another for a `while` loop iterating through a list of strings, using `continue` in both cases as instructed.



```python
print("### Using `continue` with Strings")

# Example 1: Using continue in a for loop with a string
print("\nFor loop with continue and string:")
my_string = "programming"
for char in my_string:
  print(f"Processing character: {char}")
  if char in 'aeiou':  # Check if the character is a vowel
    print(f"  Skipping vowel: {char}")
    continue  # Skip the rest of the current iteration
  print(f"  Character after continue (not a vowel): {char}")
print("For loop finished.")

# Example 2: Using continue in a while loop with a list of strings
print("\nWhile loop with continue and list of strings:")
words = ["apple", "banana", "cherry", "date", "fig"]
index = 0
while index < len(words):
  current_word = words[index]
  print(f"Processing word: {current_word}")
  if current_word == "cherry":
    print(f"  Skipping word: {current_word}")
    index += 1 # Increment before continue to avoid infinite loop
    continue # Skip the rest of the current iteration
  print(f"  Word after continue (not 'cherry'): {current_word}")
  index += 1
print("While loop finished.")
```

## Summarize the key differences and use cases of `break` and `continue`.

### Subtask:
Highlight the main distinctions between `break` and `continue` and when to use each.


**Reasoning**:
Create a markdown cell to explain the key differences and use cases of `break` and `continue` statements, following the instructions.



```python
# This is a markdown cell
# ## Key Differences and Use Cases
#
# The primary difference between `break` and `continue` statements in Python loops is their effect on the loop's execution flow.
#
# - The `break` statement **terminates the entire loop** immediately when encountered. The program execution then moves to the first statement after the loop.
# - The `continue` statement **skips the rest of the current iteration** of the loop and moves to the next iteration. The loop's execution continues with the next item (for `for` loops) or by re-evaluating the condition (for `while` loops).
#
# ### Use Cases for `break`
#
# Use `break` when you need to exit a loop entirely based on a specific condition, regardless of whether the loop has completed iterating through all its elements or met its original termination condition. Common use cases include:
#
# - **Finding a specific item:** Stop searching once the desired item is found.
# - **Encountering an error or invalid condition:** Exit the loop if something goes wrong or an unexpected value is encountered.
# - **Implementing search algorithms:** Terminate the search when the target element is located.
#
# ### Use Cases for `continue`
#
# Use `continue` when you want to skip processing for a particular iteration based on a condition but continue with the rest of the loop's iterations. Common use cases include:
#
# - **Skipping invalid data or inputs:** Process only valid entries and skip the rest.
# - **Filtering elements:** Perform operations only on elements that meet certain criteria, skipping those that don't.
# - **Optimizing loop performance:** Avoid executing unnecessary code within an iteration.
#
# ### Summary
#
# In essence, `break` is for **stopping** the loop, while `continue` is for **skipping** the current iteration and moving to the next. Choose `break` when the task is done or impossible to continue within the loop. Choose `continue` when you need to skip a specific case but want to keep processing other cases in the loop.
```
