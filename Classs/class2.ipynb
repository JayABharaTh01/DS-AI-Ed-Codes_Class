```python
# lstrip() removes leading whitespace
x = "   data"
print(x.lstrip())
```

```python
# rstrip() removes trailing whitespace
x = "data   "
print(x.rstrip())
```

```python
# strip() removes both leading and trailing whitespace
x = "   data   "
print(x.strip())
```

```python
# zfill() pads the string with zeros on the left to fill the specified width
x = "data"
print(x.zfill(8))
```

```python
# ljust() left-justifies the string within the specified width, padding with spaces
x = "data"
print(x.ljust(8), "science")
```

```python
# rjust() right-justifies the string within the specified width, padding with spaces
x = "data"
print(x.rjust(8))
```

```python
# isspace() returns True if all characters in the string are whitespace, False otherwise
x = "   "
print(x.isspace())
y = "data"
print(y.isspace())
```

```python
# find() returns the lowest index of the substring if it is found, otherwise -1
x = "data science"
print(x.find("science"))
print(x.find("Python"))
```
